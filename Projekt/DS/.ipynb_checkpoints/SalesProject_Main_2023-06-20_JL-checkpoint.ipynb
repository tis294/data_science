```python
def fill_nan_with_common_value(df, nan_col, ref_col):
    # Step 1: Plot initial relplot
    #sns.heatmap(pd.get_dummies(df[nan_col, rel_col]).corr, annot=True)

    # Step 2: Check for NaN values in 'nan_col' and 'ref_col'
    nan_rows = df[df[nan_col].isna()]
    nan_rows_ref_col_values = nan_rows[ref_col].unique()

    # Step 3: Get the most common value in nan_col (e.g. 'Nation') for each ref_col (e.g. 'premium_Country')
    common_nations = {}
    for value in nan_rows_ref_col_values:
        mode_values = df[df[ref_col] == value][nan_col].mode()
        if not mode_values.empty:
            common_nation = mode_values.iloc[0]
            common_nations[value] = common_nation

    # Step 4: Fill NaN values in 'Nation' with the most common value for the corresponding 'premium_Country'
    for value in common_nations:
        df.loc[(df[ref_col] == value) & (df[nan_col].isna()), nan_col] = common_nations[value]

    # Step 5: Plot histplot after filling NaN values
    sns.histplot(data=df, x=nan_col, y=ref_col)
```

```python
fill_nan_with_common_value(df, 'Nation_region', 'premiumCountry_region')
```
